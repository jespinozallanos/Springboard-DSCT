```python
#DEFINING TEST DATA (JULY)

te = DF_copy[91:]
te
```

```python
##Specifying Training and Test data further to feed into M.L. Regressors. 

#Training models with April, May and June data points.
#Testing models with July data points.

#Training and Testing data further split up into x and y components.

#x is all data without value_activityCalories
#y is all data with only value_activityCalories

a = tr.drop('value_activityCalories', axis = 1)  #APRIL to JUNE values
b = tr['value_activityCalories']

c = te.drop('value_activityCalories', axis = 1) #JULY values
d = te['value_activityCalories']

trainx = a
trainy = b

testx = c
testy = d
```

```python
print(DF_copy.astype(int).describe())
```

```python
##Training Linear Regression (L.R.) Model

from sklearn.linear_model import LinearRegression
from sklearn import linear_model

lm = linear_model.LinearRegression().fit(trainx, trainy)
lm
```

```python
#Predicting on training data
lm_train_predictions =  lm.predict(trainx)
lm_train_predictions
```

```python
#Predicting on test data
lm_test_predictions = lm.predict(testx)
lm_test_predictions
```

```python
#This is off of Training Data

#Actual and Predictions
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width=450, plot_height=450, title="L.R.: APRIL, MAY & JUNE: Actual (blue) & Predictions (red)")


p.line(list(range(91)), trainy, line_width=2) #Actual

p.line(list(range(91)), lm_train_predictions, color="firebrick", line_width=2) #Predictions

p.xaxis.axis_label = "Day #"
p.yaxis.axis_label = "value_activityCalories"

show(p)
```

```python
#Training set RMSE

from sklearn.metrics import mean_squared_error
import numpy as np

calories_predictions = lm_train_predictions
lin_mse = mean_squared_error(trainy, calories_predictions)
tr_lin_rmse = np.sqrt(lin_mse)
tr_lin_rmse
```

```python
#Training set Residuals

# residuals = actual label values - predicted label values

calories_predictions #predicted label values

trainy = np.array(trainy, dtype=np.int32) #actual label values

trainresid1 = (trainy) - (calories_predictions)
trainresid1
```
