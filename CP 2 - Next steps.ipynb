```python
# (a) Train the model with all May data points, except for
# the last two in the timeline (a.k.a. held-out points, a.k.a. test set).

#DF without dateTime and without value_activityCalories needed for training set.

a = DF_copy.drop('value_activityCalories', axis = 1)
b = DF_copy['value_activityCalories']

trainx = a[:29]
trainy = b[:29]

testx = a[29:]
testy = b[29:]

testy
```

```python
#Training Linear Regression (L.R.) Model

from sklearn.linear_model import LinearRegression
from sklearn import linear_model

lm = linear_model.LinearRegression().fit(trainx, trainy)
lm
```

```python
lm.predict(trainx)
```

```python
#Graph L.R. model. This is based on one set of predictions for a span of 29 days.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width=400, plot_height=400, title="Day 1- 29 Predictions")

# add a line renderer
p.line(list(range(30)), lm.predict(trainx), line_width=2)


p.xaxis.axis_label = "Day #"
p.yaxis.axis_label = "Prediction (value_activityCalories)"

show(p)
```

```python
# (b) Compute RMSE

from sklearn.metrics import mean_squared_error
import numpy as np

calories_predictions = lm.predict(trainx)
lin_mse = mean_squared_error(trainy, calories_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse #RMSE for training set
```

```python
#Compute actual residuals on training set

# residuals = actual label values - predicted label values

calories_predictions #predicted label values

trainy = np.array(trainy, dtype=np.int32) #actual label values

trainresid1 = (trainy) - (calories_predictions)
trainresid1
```

```python
# (c) Compute RMSE and actual residuals for test set

calories_predictions2 = lm.predict(testx)
lin_mse = mean_squared_error(testy, calories_predictions2)
lin_rmse = np.sqrt(lin_mse)
lin_rmse #RMSE for test set
```

```python
#actual residuals for test set

# residuals = actual label values - predicted label values

calories_predictions2 #predicted label values

testy = np.array(testy, dtype=np.int32) #actual label values

testresid1 = (testy) - (calories_predictions2)
testresid1
```
