```python
##Specifying Training and Test data further to feed into M.L. Regressors. 

#Training models with April, May and June data points.
#Testing models with July data points.

#Training and Testing data further split up into x and y components.

#x is all data without value_activityCalories
#y is all data with only value_activityCalories

a = tr.drop('value_activityCalories', axis = 1)  #APRIL to JUNE values
b = tr['value_activityCalories']

c = te.drop('value_activityCalories', axis = 1) #JULY values
d = te['value_activityCalories']

trainx = a
trainy = b

testx = c
testy = d
```

```python
print(DF_copy.astype(int).describe())
```

```python
##Training Linear Regression (L.R.) Model

from sklearn.linear_model import LinearRegression
from sklearn import linear_model

lm = linear_model.LinearRegression().fit(trainx, trainy)
lm
```

```python
#Predicting on training data
lm_train_predictions =  lm.predict(trainx)
lm_train_predictions
```

```python
#Predicting on test data
lm_test_predictions = lm.predict(testx)
lm_test_predictions
```

```python
#This is off of Training Data

#Actual and Predictions
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width=450, plot_height=450, title="L.R.: APRIL, MAY & JUNE: Actual (blue) & Predictions (red)")


p.line(list(range(91)), trainy, line_width=2) #Actual

p.line(list(range(91)), lm_train_predictions, color="firebrick", line_width=2) #Predictions

p.xaxis.axis_label = "Day #"
p.yaxis.axis_label = "value_activityCalories"

show(p)
```

```python
#Training set RMSE

from sklearn.metrics import mean_squared_error
import numpy as np

calories_predictions = lm_train_predictions
lin_mse = mean_squared_error(trainy, calories_predictions)
tr_lin_rmse = np.sqrt(lin_mse)
tr_lin_rmse
```

```python
#Training set Residuals

# residuals = actual label values - predicted label values

calories_predictions #predicted label values

trainy = np.array(trainy, dtype=np.int32) #actual label values

trainresid1 = (trainy) - (calories_predictions)
trainresid1
```

```python
#Test set RMSE

calories_predictions2 = lm_test_predictions
lin_mse = mean_squared_error(testy, calories_predictions2)
te_lin_rmse = np.sqrt(lin_mse)
te_lin_rmse
```

Conclusion: 

L.R.'s RMSE values are high, but not the highest. Also, it has a relatively small gap between its Training RMSE and Test RMSE values. 
This means that the model was trained to be general enough, minimized its cost function. Overfitting was not present. 
However, one downside to this model is that its Training RMSE value is higher than its Test RMSE value. Usually, it 
is the other way around. Though there are ways to improve this, which are discussed at the end. 

DT model has one of the lowest RMSE value, which comes from its Training data set. Though D.T.'s are known to fit its data very well. Given 
this close fit, the Test RMSE value suffered with huge RMSE value. Thus, the gap between its Training and Test set RMSE values
is huge. Though good side to this is that its Training RMSE is lower than its Test RMSE value, which is expected.

R.F. model's assessment is different in the sense that for the Training error rate, the oob_score is reported. This is compared to the Test error rate in order to determine how representative is oob_score /Training set versus how representative is Test RMSE/ Test set. If the oob_score is close to the Test RMSE values then Training set performed close to Test set. However, this is not the case. Training set did significantly better than Training set on built model. So RF is not a good model to choose to make predictions with Test set. Although, this is the case, RF's Test RMSE value is comparable to L.R.'s Training and Test RMSE values. 

In the end, L.R. wins as the best model because it has relatively the lowest RMSE values and has minimal gap between its
Training RMSE and Test RMSE values. Thus, it does not overfit its data.

Recommended solutions to resolve the issue with the Training RMSE value being greater than the Test RMSE value are:
1. Cross-validate, split/reshuffle the data differently so that the Test data is not so small
2. Have fewer parameters to reduce possible noise in Training data set
3. Constrain the model so there are less errors
4. Gather more Training data.

Lastly, it could have just been random chance that the Test data was able to fit the model better than the Training data.
