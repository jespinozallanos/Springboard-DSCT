```python
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#LOADING ALL DATA (APRIL, MAY, JUNE and JULY)
```

```python
#Loading JSON file into notebook

with open('elevation.json') as json_data:
    a = json.load(json_data)
    print(a)
```

```python
#Transforming JSON into Pandas dataframe

a = json_normalize(a, 'activities-elevation')
a
```

```python
#Loading JSON file into notebook

with open('activitycalories.json') as json_data:
    b = json.load(json_data)
    print(b)
```

```python
#Transforming JSON into Pandas dataframe

b = json_normalize(b, 'activities-activityCalories')
b
```

```python
#Loading JSON file into notebook

with open('fairlyactiveminutes.json') as json_data:
    c = json.load(json_data)
    print(c)
```

```python
#Transforming JSON into Pandas dataframe

c = json_normalize(c, 'activities-minutesFairlyActive')
c
```

```python
#Loading JSON file into notebook

with open('floors.json') as json_data:
    d = json.load(json_data)
    print(d)
```

```python
#Transforming JSON into Pandas dataframe

d = json_normalize(d, 'activities-floors')
d
```

```python
#Loading JSON file into notebook

with open('steps.json') as json_data:
    e = json.load(json_data)
    print(e)
```

```python
#Transforming JSON into Pandas dataframe

e = json_normalize(e, 'activities-steps')
e
```

```python
#Loading JSON file into notebook

with open('veryactiveminutes.json') as json_data:
    f = json.load(json_data)
    print(f)
```

```python
#Transforming JSON into Pandas dataframe

f = json_normalize(f, 'activities-minutesVeryActive')
f
```

```python
first_merge = pd.merge(a, b, on='dateTime', how='outer', suffixes=('_elevation', '_activityCalories'))
first_merge
```

```python
second_merge = pd.merge(pd.merge(first_merge,c,on='dateTime'),d,on='dateTime', suffixes=('_fairlyActiveMinutes', '_floors'))
second_merge
```

```python
final_merge = pd.merge(pd.merge(second_merge,e,on='dateTime'),f,on='dateTime', suffixes=('_steps', '_veryActiveMinutes'))
final_merge
```

```python
#Need to make a copy of main DF to make table without dateTime to feed into M.L Regressors.

DF_copy = final_merge.copy()
DF_copy.tail()
```

```python
#This bracket will be the TRAINING data. 

DF_copy[:91]
```

```python
#This bracket will be the TEST data.
DF_copy[91:]
```

```python
del DF_copy['dateTime']
```

```python
#This is used to create training set and test set.

DF_copy.tail() #No dateTime
```

```python
#DEFINING TRAINING DATA (APRIL, MAY and JUNE)

tr = DF_copy[:91]
tr
```

```python
#DEFINING TEST DATA (JULY)

te = DF_copy[91:]
te
```

```python
##Specifying Training and Test data further to feed into M.L. Regressors. 

#Training models with April, May and June data points.
#Testing models with July data points.

#Training and Testing data further split up into x and y components.

#x is all data without value_activityCalories
#y is all data with only value_activityCalories

a = tr.drop('value_activityCalories', axis = 1)  #APRIL to JUNE values
b = tr['value_activityCalories']

c = te.drop('value_activityCalories', axis = 1) #JULY values
d = te['value_activityCalories']

trainx = a
trainy = b

testx = c
testy = d
```

```python
print(DF_copy.astype(int).describe())
```

```python
##Training Linear Regression (L.R.) Model

from sklearn.linear_model import LinearRegression
from sklearn import linear_model

lm = linear_model.LinearRegression().fit(trainx, trainy)
lm
```

```python
#Predicting on training data
lm.predict(trainx)
```

```python
#Predicting on test data
lm.predict(testx)
```

```python
#This is off of Training Data

#Actual and Predictions
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width=450, plot_height=450, title="L.R.: APRIL, MAY & JUNE: Actual (blue) & Predictions (red)")


p.line(list(range(91)), trainy, line_width=2) #Actual

p.line(list(range(91)), lm.predict(trainx), color="firebrick", line_width=2) #Predictions

p.xaxis.axis_label = "Day #"
p.yaxis.axis_label = "value_activityCalories"

show(p)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

# Create traces
trace0 = go.Scatter(
    x = list(range(91)),
    y = trainy,
    mode = 'actual',
    name = 'actual'
)
trace1 = go.Scatter(
    x = list(range(91)),
    y = lm.predict(trainx),
    mode = 'predictions',
    name = 'predictions'
)

data = [trace0, trace1]

layout = dict(title = 'L.R.: April, May and June',
              xaxis = dict(title = 'Day #'),
              yaxis = dict(title = 'value_activityCalories'),
              )

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='styled-line')
```

```python
#Training set RMSE

from sklearn.metrics import mean_squared_error
import numpy as np

calories_predictions = lm.predict(trainx)
lin_mse = mean_squared_error(trainy, calories_predictions)
tr_lin_rmse = np.sqrt(lin_mse)
tr_lin_rmse
```

```python
#Training set Residuals

# residuals = actual label values - predicted label values

calories_predictions #predicted label values

trainy = np.array(trainy, dtype=np.int32) #actual label values

trainresid1 = (trainy) - (calories_predictions)
trainresid1
```

```python
#Test set RMSE

calories_predictions2 = lm.predict(testx)
lin_mse = mean_squared_error(testy, calories_predictions2)
te_lin_rmse = np.sqrt(lin_mse)
te_lin_rmse
```

```python
#Test set Residuals

# residuals = actual label values - predicted label values

calories_predictions2 #predicted label values

testy = np.array(testy, dtype=np.int32) #actual label values

testresid2 = (testy) - (calories_predictions2)
testresid2
```

```python
#Conclusion: 

# Based on RMSE values alone, D.T. model looks promising since it has a value of 16.0632205332. This is the lowest value in the whole table.
# However, when comparing its Training and Test values, there is a huge gap. This means overfitting occurred,
# which makes the model a poor one.

# Another model that overfit was the R.F. model. A lot may have to do with the model training on too large of a a tree,
# which reduces the model being general enough to predict well on new test data. Though compared to D.T. model, this one is slightly better
# because the gap between the Training and Test RMSE's are not as huge. 

# When looking at the remaining model, we see that the L.R. model does not have as low as RMSE value as the D.T. model, but the L.R. model has a smaller gap between 
# Training RMSE and Test RMSE values compared to both D.T. and R.F. models.
# So due to having close RMSE values between Tranining and Test sets and having some relatively low values to all values in the table,
# the best model is the L.R. one. 
```
