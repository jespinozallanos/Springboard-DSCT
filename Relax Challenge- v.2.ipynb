```python
#Sort "engage" by users' number of visits

#Output user_id's that have cumulative =>3 visited

visits = engage["user_id"].value_counts()
adopted = visits[visits >= 3]
 
adopted
#Based on this, can see that out of 207917 time stamps, there are 2248 users who have logged in 
#to their device 3 or more instances. 

#Need to see how spread apart are these log-ins...7 days or more?
```

```python
#Breakdown:

# 1. add "adopted user" label to each row that fits criteria
# (new df has both adopted_user = 1 and adopted_user = NaN--> 0
# 2. drop irrelevant columns, leaving only "object_id", "opted_in_to_mailing_list" and "enabled_for_marketing_drip"
# 3. set up trainx and trainy
# 4. feed these into .fit to build M.L. models
# 5. .predict to predict on training set and test set
```

```python
#1

for index, val in adopted.iteritems(): 
    users.loc[users['object_id'] == index, 'adopted_user'] = 1 #index is used to connect with users df 'object_id' column
    
#Check that transformation went through with confirmed adopted user
users.loc[users['object_id'] == 3623]

#Checked
```

```python
users.head()
```

```python
#Drop all NaN values in order to properly train M.L. Models 

users = users.fillna(0)
users
```

```python
#What makes them an "adopted user"?

#Now that it has been determined which users are "adopted user's", need to figure out 
#what factors into multiple log-ins into devices.

#Some features that will be considered for this model will be: 

#'creation_source', 'opted_in_to_mailing_list' & 'enabled_for_marketing_drip'

# creation_source tells how 'adopted user' would first get onboard and 'opted_in_to_mailing_list' & 
#'enabled_for_marketing_drip' show their level of commitment and therefore their continual usage
```

```python
# 2. drop irrelevant columns, leaving only "object_id", "opted_in_to_mailing_list" and "enabled_for_marketing_drip"

users.drop(['creation_time', 'name', 'email', 'creation_source', 'last_session_creation_time', 'org_id', 'invited_by_user_id'], axis=1, inplace=True)
users
```
